**answer4**:- the star schema is  better to use than the snowflake schema. It is the officially recommended approach by Microsoft for its simplicity, fast query execution, and user-friendliness.
reasons  to use star schema are
1.Faster Performance: Fewer joins are required between tables, allowing Power BI's engine to process filters and visuals much faster.
2. Better Compression: Dimension tables are left flat (denormalized), which compress extremely well in Power BI's memory.
3. User-Friendly: Business users building their own reports can understand the structure immediately.
